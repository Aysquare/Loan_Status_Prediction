### Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

### Data Collection and Preprocessing

```python
# Loading the dataset into pandas DataFrame
loan_dataset = pd.read_csv('C:\\Users\\USER\\Desktop\\Kaggle\\loan status.csv')
```

```python
### Investigating the first 5 rows of the DataFrame

loan_dataset.head()
```

```python
# Investigating the number of rows and columns
loan_dataset.shape
```

```python
# Investigating the number of missing values in each column
loan_dataset.isna().sum()
```

```python
# Statistical measure of the dataset

loan_dataset.describe()
```

```python
# Dropping the missing values

loan_dataset = loan_dataset.dropna()
loan_dataset.isna().sum()
```

```python
# Encoding the categorical variables

loan_dataset.replace({'Loan_Status': {'Y': 1, 'N': 0}}, inplace = True)
```

```python
loan_dataset.head()
```

### Investigating the distribution of the numerical dataset

```python
# Dependent column value

print(loan_dataset['Dependents'].value_counts())
```

```python
# Replacing 3+ with 4

loan_dataset = loan_dataset.replace(to_replace = '3+', value = 4)
```

```python
print(loan_dataset['Dependents'].value_counts())
```

### Data Visualization

Investigating the distribution of various feature variables with respect to loan status

### Label Encoding

```python
# Converting Categorical columns to numerical value

loan_dataset.replace({'Married': {'No': 0, 'Yes':1}, 'Gender': {'Male':1, 'Female': 0}, 'Self_Employed': {'No': 0, 'Yes': 1},'Property_Area': {'Semiurban': 1, 'Rural': 0, 'Urban': 2}, 'Education': {'Graduate': 1, 'Not Graduate': 0}}, inplace = True)
```

```python
# Investigating the first 5 rows to confirm if the changes have been effected

loan_dataset.head()
```

```python
# Seperating the feature and target variable

X = loan_dataset.drop(['Loan_ID', 'Loan_Status'],axis = 1)
Y = loan_dataset['Loan_Status']
```

```python
print(X)
```

```python
print(Y)
```

### Train Test Split

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 2)
```

### Training the model: 

Support Vector Machine

```python
classifier = svm.SVC(kernel = 'linear')
classifier.fit(X_train, Y_train)
```

### Model Evaluation

Train data

```python
train_data_pred = classifier.predict(X_train)

accuracy = accuracy_score(Y_train, train_data_pred)
print('Accuracy score of train data: ', accuracy)
```

Test data

```python
test_data_pred = classifier.predict(X_test)

accuracy = accuracy_score(Y_test, test_data_pred)
print('Accuracy score of test data: ', accuracy)
```
